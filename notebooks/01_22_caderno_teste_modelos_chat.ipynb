```python
from jupyter_ai_persona_manager import BasePersona, PersonaDefaults
from jupyterlab_chat.models import Message
from ollama import Client
import os

OLLAMA_HOST = os.environ.get("OLLAMA_HOST", "http://ollama:11434")
client = Client(host=OLLAMA_HOST)


class OllamaPersona(BasePersona):

    @property
    def defaults(self):
        return PersonaDefaults(
            name="Ollama Local",
            description="Assistente local usando Ollama no JupyterLab",
            system_prompt="Você é um assistente técnico em português para Python, JupyterLab e análise de dados."
        )

    async def process_message(self, message: Message):
        response = client.chat(
            model="llama3.2:3b",
            messages=[
                {
                    "role": "system",
                    "content": "Responda em português, com explicações claras e código quando necessário."
                },
                {
                    "role": "user",
                    "content": message.body
                }
            ]
        )

        self.send_message(response.message.content)
```

```python
from ollama import Client

OLLAMA_HOST = "http://ollama:11434"

client = Client(host=OLLAMA_HOST)

# listar modelos
modelos = client.list()
print("Modelos disponíveis:")

for m in modelos["models"]:
    print("-", m["model"])

# escolha um modelo da lista
modelo = modelos["models"][0]["model"]

print("\nUsando modelo:", modelo)

response = client.chat(
    model=modelo,
    messages=[
        {
            "role": "user",
            "content": "Explique em português o que é JupyterLab."
        }
    ]
)

print(response.message.content)
```

```python
from ollama import Client

client = Client(host="http://ollama:11434")

modelo = "qwen2.5-coder:14b"  # recomendado para Python/Jupyter

response = client.chat(
    model=modelo,
    messages=[
        {
            "role": "system",
            "content": "Você é um assistente técnico em português especializado em Python, JupyterLab e análise de dados."
        },
        {
            "role": "user",
            "content": "Explique em português o que é JupyterLab."
        }
    ],
    options={
        "temperature": 0.2,
        "num_predict": 800
    }
)

print(response.message.content)
```
